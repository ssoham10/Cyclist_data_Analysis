```python
import pandas as pd
import datetime as dt
import statistics
```

```python
dataset = pd.read_csv("cyclist_counts_copy.csv")
```

```python
dataset[['date', 'time']] = dataset[['date', 'time']].apply(pd.to_datetime)
```

```python
dataset
```

```python
dataset.dtypes
```

```python
without_date = dataset['time'].apply( lambda d : d.time() )
dataset['time'] = without_date
```

```python
dataset
```

```python
dataset.dtypes
```

```python
month_df= (dataset.groupby([dataset['date'].dt.month.rename('Month')])
          [['fleher deich ost stromaufwaerts', 'fleher deich west stromabwaerts', 'okb nord', 'okb sued']].sum()
         .reset_index())
```

```python
month_df
```

```python
columns = ['fleher deich ost stromaufwaerts', 'fleher deich west stromabwaerts', 'okb nord', 'okb sued' ]
for col in columns:
  print("\nFor column: " + col)
  print('Mean: ' + str(statistics.mean(month_df[col])))
  print('Median: ' + str(statistics.median(month_df[col])))
  #print('Mode: ' + str(statistics.mode(month_df[col])))
  print('Range: '+ str(month_df[col].max() - month_df[col].min()))
  print('Standard Deviance: ' + str(statistics.stdev(month_df[col])))
  print('Variance: ' + str(statistics.variance(month_df[col])))
```
